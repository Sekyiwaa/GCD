```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exact_vs_asymptotic(lower_bound,upper_bound):
  x=np.linspace(lower_bound,upper_bound)
  y=(3 * np.square(x)) + (200 * x) + 31
  z = 3*x*x
  fig = plt.figure(figsize = (6, 4))
  plt.title(f"Exact cost vs Asymptotic cost with range {lower_bound,upper_bound}")
  plt.xlabel('n : Size of the problem')
  plt.ylabel('T(n) :Total number of instructions')
  plt.plot(x, y, label="T(n)= 3n^2 + 200n + 31", color="purple")
  plt.plot(x, z, label="O(T(n))= 3n^2", color="green")
  plt.grid(True)
  plt.legend()
```

```python
exact_vs_asymptotic(1,100)
```

```python
exact_vs_asymptotic(1,1000)
```

```python
exact_vs_asymptotic(1,10000)
```
